```python
# Importar las librerias que vamos a utilizar / Importing the libraries we are going to use
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Leer el dataset amazon.csv / Read the dataset amazon.csv
amazon_df = pd.read_csv('amazon.csv')

# Mostrar las primeras filas del dataset / Show the first rows of the dataset
amazon_df.head()
```

```python
# Mostrar informacion sobre el dataset / Show information about the dataset
amazon_df.info()
```

```python
# Eliminar las columnas innecesarias / Remove unnecessary columns
amazon_df = amazon_df.drop(columns=['Unnamed: 3', 'about_product', 'review_title', 'review_content', 'img_link', 'product_link'])
```

```python
# Eliminar campos vacios / Remove empty fields
amazon_df = amazon_df.dropna()

# Eliminar duplicados / Remove duplicates       
amazon_df = amazon_df.drop_duplicates()

# reset index / Reset index
amazon_df = amazon_df.reset_index(drop=True)

# mostrar informacion sobre el dataset / Show information about the dataset
amazon_df.info()
```

```python
# Mostrar informacion aleatoria del dataset / Show random information of the dataset
amazon_df.sample(15)
```

```python
# Cortar las categorias hasta '|' / Cut the categories until '|'
amazon_df['category'] = amazon_df['category'].str.split('|').str[0]

#Editar las categorias / Edit the categories    
amazon_df['category'] = amazon_df['category'].replace({'Computers&Accessories': 'Computers & Accessories', 'MusicalInstruments': 'Musical Instruments', 'OfficeProducts': 'Office Products', 'Home&Kitchen': 'Home & Kitchen', 'HomeImprovement': 'Home Improvement', 'Toys&Games': 'Toys & Games', 'Car&Motorbike': 'Car & Motorbike', 'Health&PersonalCare': 'Health & Personal Care'})


# Mostrar las categorias / Show the categories
print(amazon_df['category'].unique())
```

```python
# Convertir columnas necesarias a tipos numéricos / Convert necessary columns to numeric types
amazon_df['discounted_price'] = amazon_df['discounted_price'].replace('[₹,]', '', regex=True).astype(float)
amazon_df['actual_price'] = amazon_df['actual_price'].replace('[₹,]', '', regex=True).astype(float)
amazon_df['discount_percentage'] = amazon_df['discount_percentage'].str.replace('%', '').astype(float)
amazon_df['rating'] = pd.to_numeric(amazon_df['rating'], errors='coerce')
amazon_df['rating_count'] = amazon_df['rating_count'].str.replace(',', '').str.replace(' ', '')
amazon_df['rating_count'] = pd.to_numeric(amazon_df['rating_count'], errors='coerce')

```

```python
#guardar el dataset limpio / Save the clean dataset
amazon_df.to_csv('amazon_clean.csv', index=False)
```

# Descuento promedio por categoria

```python
# Descuento promedio por categoria / Average discount by category
discount_avg = amazon_df.groupby('category')['discount_percentage'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=discount_avg.values, y=discount_avg.index)
plt.title('Promedio de descuento por categoría')
plt.xlabel('Descuento (%)')
plt.ylabel('Categoría')
plt.tight_layout()
plt.show()
```

# Calificacion promedio por categoria

```python
# Calificacion promedio por categoria / Average rating by category
rating_avg = amazon_df.groupby('category')['rating'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=rating_avg.values, y=rating_avg.index)
plt.title('Promedio de calificación por categoría')
plt.xlabel('Rating promedio')
plt.ylabel('Categoría')
plt.tight_layout()
plt.show()
```

# Productos con mayor numero de reseñas

```python
# Productos con más reseñas / Products with the most reviews
top_rated = amazon_df.sort_values(by='rating_count', ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_rated['rating_count'], y=top_rated['product_name'], palette='crest')
plt.title('Productos con más reseñas')
plt.xlabel('Cantidad de reseñas')
plt.ylabel('Producto')
plt.tight_layout()
plt.show()
```

# Productos con mayor descuento

```python
# Producto con mas descuento / Product with the most discount
top_discounted = amazon_df.sort_values(by='discount_percentage', ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_discounted['discount_percentage'], y=top_discounted['product_name'])
plt.title('Producto con más descuento')
plt.xlabel('Descuento (%)')
plt.ylabel('Producto')
plt.tight_layout()
plt.show()
```

# ¿Relación entre descuento y calificacion ?

```python
# Relación entre descuento y rating / Relationship between discount and rating
plt.figure(figsize=(8,6))
sns.scatterplot(data=amazon_df, x='discount_percentage', y='rating')
plt.title('Relación entre Porcentaje de Descuento y Calificación')
plt.xlabel('Porcentaje de Descuento (%)')
plt.ylabel('Rating')
plt.grid(True)
plt.show()

# También puedes calcular la correlación:
correlacion = amazon_df['discount_percentage'].corr(amazon_df['rating'])
print(f"Correlación entre descuento y rating: {correlacion:.2f}")


```

Hay una relación débilmente negativa entre el porcentaje de descuento y la calificación de los productos.

Interpretación práctica:
➔ A medida que el descuento aumenta, la calificación tiende a bajar ligeramente, pero la relación es muy débil.
➔ No podemos decir que ofrecer un gran descuento esté fuertemente asociado a tener malas calificaciones — simplemente hay una leve tendencia.

Productos con grandes descuentos podrían ser de menor calidad o menos valorados originalmente.

O bien, los productos de alta calidad no necesitan ser fuertemente descontados para venderse.

También puede ser ruido: muchos factores externos afectan las calificaciones, no solo el descuento.
